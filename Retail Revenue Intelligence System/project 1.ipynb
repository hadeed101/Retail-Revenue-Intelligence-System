### **Retail Revenue Intelligence System**

```python
import pandas as pd
import numpy as np
from pandas.conftest import ascending

```

```python
df =pd.read_csv("Retail Revenue Intelligence System/project 001.csv")
```

```python
df
```

**Total Revenue**
* Create a new column name Revenue
* calculate the total sales of the company

```python
df['Total Revenue']=[0 for i in range(len(df))]
```

```python
df
```

New column Created "Total Revenue"

```python
def fx(a):
    return a
df['Total Revenue']=df['quantity']*df['price'].apply(fx)
```

```python
df
```

**Performance as per Region**

```python
df.groupby('region')['Total Revenue'].sum()
```

* NOrth region is doing well
* East should be focused.

#### **Product Ranking**

```python
df.groupby('product')['Total Revenue'].sum().sort_values(ascending=False).head(5)
```

Top Revenue Generator Products.

#### **Category Analysis**

we have to check our total revenue by catogery , so we can check profitable & Non- profitable Categories so we can focus on those categories

```python
df.groupby('category')['Total Revenue'].sum()
```

* Top Profitable Category, Sports.
* Least Profitable Category ,Clothing.

#### **Time Analysis**

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.groupby(df['order_date'].dt.to_period('M'))['Total Revenue'].sum().sort_values(ascending=False)
```

* Mostly sales are in August.
* lowest sales are in July.
